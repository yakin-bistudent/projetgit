```python
x=[0.25,0.25,1.25,0.5,1,0.25,0.6,0,-0.6,-0.25,-1,-0.5,-1.25,-0.25,-0.25,0.25]
y=[0,0.5,0.5,1,1,1.5,1.5,2,1.5,1.5,1,1,0.5,0.5,0,0]
```

```python
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel(" x")
plt.ylabel("y")
plt.title("sapin de noel")
plt.fill_between(x,y,color="green")
plt.scatter(x,y,color="pink",marker="*",s=200)
plt.show()
```
